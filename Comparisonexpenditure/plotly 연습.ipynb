```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mlp
import seaborn as sns
import plotly.plotly as py
import cufflinks as cf
import plotly.offline as pyo
import plotly.graph_objs as go
cf.go_offline(connected=True)
import chart_studio
chart_studio.tools.set_credentials_file(username='username', api_key='api_key')
import chart_studio.plotly as py
```

```python
data = pd.read_excel('2019 월별 전국 관광지 지출액.xlsx')
data
```

```python
data1 = pd.read_excel('2020 월별 전국 관광지 지출액.xlsx')
data1
```

```python
data1["렌터카 전년 대비 증가율"] = (data1["렌터카"]-data["렌터카"]) / data["렌터카"] * 100
data1
```

```python
trace1 = go.Scatter(x=data["월"], y=data["합계"] , name="2019 총 지출액")
trace2 = go.Scatter(x=data1["월"], y=data1["합계"], name='2020 총 지출액')
totaldata = [trace1, trace2]
layout = go.Layout(title='전국 관광지 지출액 비교')
fig = go.Figure(data=totaldata, layout=layout)
pyo.iplot(fig)
```

```python
trace3 = go.Bar(x=data1["월"], y=data1["렌터카 전년 대비 증가율"], name='렌터카 지출액 증가율')
totaldata = [trace3]
layout = go.Layout(title='렌터카 전년 대비 증가율')
fig = go.Figure(data=totaldata, layout=layout)
pyo.iplot(fig)
```
