```python
data2019 = pd.read_excel("./2019 일별 방문자수/2019 일별 방문자수 전국.xlsx", index_col=[0,1])
data2020 = pd.read_excel("./2020 일별 방문자수/2020 일별 방문자수 전국.xlsx", index_col=[0,1])
data1 = pd.read_excel("./2019 일별 방문자수/전국 전체 월별 방문자수 2019.xlsx")
data2 = pd.read_excel("./2020 일별 방문자수/전국 전체 월별 방문자수 2020.xlsx")
```

```python
visitor2019_1 = go.Scatter(x=data2019.index[:,0], y=data2019["현지인(a)"], name='현지인')
visitor2019_2 = go.Scatter(x=data2019.index[:,0], y=data2019["외부방문자(b+c)"], name='외부방문자')
# visitor2 = go.Scatter(x=data1_2.index,)
totaldata = [visitor2019_1,visitor2019_2]
layout = go.Layout(title='2019년 일별 전국 방문자 수 데이터')
fig = go.Figure(data=totaldata, layout=layout)
pyo.iplot(fig)
```

```python
visitor2020_1 = go.Scatter(x=data2020.index.name, y=data2020["현지인(a)"], name='현지인')
visitor2020_2 = go.Scatter(x=data2020.index.name, y=data2020["외부방문자(b+c)"], name='외부방문자')
# visitor2 = go.Scatter(x=data1_2.index,)
totaldata = [visitor2020_1,visitor2020_2]
layout = go.Layout(title='2020년 일별 전국 방문자 수 데이터')
fig = go.Figure(data=totaldata, layout=layout)
pyo.iplot(fig)
```
