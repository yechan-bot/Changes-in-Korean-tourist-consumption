```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mlp
import seaborn as sns
import plotly.plotly as py
import cufflinks as cf
import plotly.offline as pyo
import plotly.graph_objs as go
cf.go_offline(connected=True)
```

```python
data = pd.read_excel('2018 월별 전국 목적지 검색 횟수.xlsx')
data
```

```python
data1 = pd.read_excel('2019 월별 전국 목적지 검색 횟수.xlsx')
data1
```

```python
data2 = pd.read_excel('2020 월별 전국 목적지 검색 횟수.xlsx')
data2.columns
```

```python
trace1 = go.Scatter(x=data["월"], y=data["전국"], name="2018 목적지 검색 횟수")
trace2 = go.Scatter(x=data1["월"], y=data1["전국"], name='2019 목적지 검색 횟수')
trace3 = go.Scatter(x=data2["월"], y=data2["전국"], name='2020 목적지 검색 횟수')
totaldata = [trace1,trace2, trace3]
layout = go.Layout(title='전국 목적지 검색 횟수 비교')
fig = go.Figure(data=totaldata, layout=layout)
pyo.iplot(fig)
```
