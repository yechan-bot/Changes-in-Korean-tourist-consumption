```python
data0 = pd.read_excel("./2018~2020 목적지 검색 횟수 비교/2020 월별 전국 목적지 검색 횟수.xlsx")
data0
```

```python
data1 = pd.read_excel("전국 전체 월별 방문자수 2020.xlsx")
data1
```

```python
data2 = pd.read_excel("./2019~2020 지출액 비교/2020 월별 전국 관광지 지출액.xlsx")
data2
```

```python
search2020 = data0["전국"]
visitor2020 = data1["합계"]
price2020 = data2["합계"]
```

```python
data= pd.DataFrame({"search" : search2020, "visitor" :  visitor2020, "price" : price2020})
data
```

```python
corr = data.corr(method='pearson')
corr
```

```python
dataheatmap = sns.heatmap(corr, cbar = True, annot = True, annot_kws={'size' : 20}, fmt = '.2f', square = True, cmap='Purples')
```

```python
sns.pairplot(data, kind='reg', height=4)
```
