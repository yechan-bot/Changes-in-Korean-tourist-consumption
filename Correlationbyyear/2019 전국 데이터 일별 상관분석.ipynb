```python
data0 = pd.read_excel("광역지자체 일별 이동통신 데이터 목적지 검색 2019.xlsx")
data0
```

```python
data1 = pd.read_excel("전국 전체 일별 방문자수 2019.xlsx")
data1
```

```python
data2 = pd.read_excel("광역지자체 일별 관광지출액 2019.xlsx")
data2
```

```python
search2019 = data0["전국"]
visitor2019 = data1["외부방문자(b+c)"]
price2019 = data2["전국"]
```

```python
data= pd.DataFrame({"search" : search2019, "visitor" :  visitor2019, "price" : price2019})
data
```

```python
corr = data.corr(method='pearson')
corr
```

```python
dataheatmap = sns.heatmap(corr, cbar = True, annot = True, annot_kws={'size' : 20}, fmt = '.2f', square = True, cmap='Oranges')
```

```python
sns.pairplot(data, kind='reg', height=4) 
```
